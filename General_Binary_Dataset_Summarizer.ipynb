```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load the CSV file into a Pandas DataFrame
file_path = input("Please enter the file path of your CSV file: ")
df = pd.read_csv(file_path)

# Prompt user for column name of the text data
text_col_name = input("Please enter the name of the text data column: ")

# Prompt user for the column name of the binary labels
label_col_name = input("Please enter the name of the binary label column (1 represents the category of interest): ")

# Count the number of words in each text data
df["num_words"] = df[text_col_name].apply(lambda x: len(x.split()))

# Subset the data by class
ham_data = df[df[label_col_name] == 0]
spam_data = df[df[label_col_name] == 1]

# Create a box plot using seaborn
fig, ax = plt.subplots(figsize=(8,6))
sns.boxplot(x=label_col_name, y="num_words", data=df)
plt.xlabel("Class", fontsize=16)
plt.ylabel("Number of Words", fontsize=16)
plt.title("Distribution of Words per Text Data", fontsize=18)
plt.show()

# Overlay word frequency distributions
fig, ax = plt.subplots(figsize=(8,6))
sns.kdeplot(ham_data["num_words"], label="ham")
sns.kdeplot(spam_data["num_words"], label="spam")
plt.xlabel("Number of Words", fontsize=16)
plt.ylabel("Frequency", fontsize=16)
plt.title("Word Frequency Distribution by Class in " + file_path.split('/')[-1], fontsize=18)
plt.legend(fontsize=14)
plt.show()

# Show word frequency distribution by class in separate histograms
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12, 6))
sns.histplot(ham_data["num_words"], bins=50, ax=ax1, color="#4CB391")
ax1.set(title="Ham Text Data", xlabel="Number of Word Tokens", ylabel="Frequency")
sns.histplot(spam_data["num_words"], bins=50, ax=ax2, color="#E36F5A")
ax2.set(title="Spam Text Data", xlabel="Number of Word Tokens", ylabel="Frequency")
plt.suptitle("Word Token Count Distribution by Class in " + file_path.split('/')[-1], fontsize=18)
plt.show()

# Count the number of rows in each label category
num_1 = df[label_col_name].sum()
num_0 = len(df) - num_1

# Generate summary table
grouped = pd.DataFrame({
    "Dataset": [file_path.split('/')[-1]],
    "Class 0 Count": [num_0],
    "Class 1 Count": [num_1],
    "Ratio of Class 1 Entries": [num_1 / len(df)],
})
print(tabulate(grouped, headers='keys', tablefmt='fancy_grid'))
```
